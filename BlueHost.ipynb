```python
import requests
from bs4 import BeautifulSoup as Soup
from fake_useragent import UserAgent
us = UserAgent()
import pandas as pd
```

```python
def BLueHost():
    search = input("Enter Domain: ")
    html = requests.get(f"https://registration.bluehost.com/domains/search/{search}.com?suggestions=100&propertyID=52",headers={"User-Agent":f"{us.random}"}).json()
    prod_domain = []
    for i in range(len(html['suggestions'])):
        val = html['suggestions'][i]['domainInfo']['domain']
        prod_domain.append(val)
    prod_price = []
    for i in range(len(html['suggestions'])):
        val = "$"+html['suggestions'][i]['terms'][0]['basePrice']
        prod_price.append(val)
    prod_tld = []
    for i in range(len(html['suggestions'])):
        val = html['suggestions'][i]['domainInfo']['tld']
        prod_tld.append(val)
    data = {
        "Domain Name":prod_domain,
        "Tld":prod_tld,
        "Price":prod_price
    }
    df = pd.DataFrame(data)
    df.sort_values('Domain Name')
    df['Provider'] = "Bluehost"
    return df
```
